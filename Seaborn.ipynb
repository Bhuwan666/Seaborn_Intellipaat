```python
# the ways of styling themes are as follows:
# white
# dark
# whitegrid
# darkgrid
# ticks

# set_context() allows us to override default parameters. this affects things like the size of the labels, lines, and other elements of the plot, but not the overall style -> set_style(). the context are:
# poster
# paper - Ideal for pricing in academic papers
# notebook - Good for working inside Jupyter notebooks (default setting).
# Talk - Good for presentation slides where readability matters.
```

```python
sns.set_style("ticks")
sns.set_context("paper")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue = "island", palette = "Dark2")
sns.despine()  # removes top and right spine
```

```python
# palette

# Set1 - bright
# Set2 - softer
# Set3 - mix

# deep
# muted
# bright
# pastel
# dark
# colorblind

# palette type

# Qualitative - distinguishing categories or groups
# Sequential - showing magnitude or intensity, like population or revenue
# Diverging - showing positive vs negative change 

```

```python
sns.heatmap(data = penguins[columns].corr(), annot = True, cmap = "Blues", vmin = -0.2,linewidth = "2", linecolor = "black")   #  negative scale start from -0.2, changed the scale change
```

```python
# Rug plot :

sns.rugplot(data = penguins, x = "body_mass_g", hue = "species", palette = "Set2") 
```

```python
sns.rugplot(data = penguins, y = "body_mass_g", hue = "species", palette = "Set2", height = 0.2) 
```

```python
# pair plot :

sns.pairplot(data = penguins[columns])   # filter categorical column and works on numerical columns automatically
# diagonal plots are of histogram and upper and lower triagles are scatterplot
```

```python
sns.pairplot(data = penguins, hue = "sex", palette = "Set2", diag_kind = "kde")   # kde plot for diagonal directional
# diagonal plots are kde 
```

```python
sns.pairplot(data = penguins, hue = "sex", palette = "Set2", diag_kind = "hist")
```

```python
# pair grid -> subplot grid for plotting pairwise relationships in a dataset

graph = sns.PairGrid(data = penguins, hue = "sex", palette = "Set2")
graph.map_upper(sns.scatterplot) #upper triangle
graph.map_lower(sns.kdeplot) # lower triangle
graph.map_diag(sns.histplot) # diagonal
```
